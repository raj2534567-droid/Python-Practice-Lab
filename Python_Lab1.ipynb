# while loops and loop control: break, continue, pass

**Exercise 1:** Use a while loop to count down from 3 to 1 and then print 'Go!'.

```python
i = 3
while i >= 1:
    print(i)
    i -=1
print("Go!")
```

**Exercise 2:** Loop from 1..10 and print only odd numbers using continue.

```python
i = 1       # initializer
while i<= 10:
    if (i% 2==0):   # condition
        i += 1
        continue   # skipping even number
    print(i)       # printing only odd number
    i += 1
```

**Exercise 3:** Use a loop to search for the first multiple of 7 between 1..50 and break when found, then print it.

```python
i = 7
while i <= 50:
    if (i % 7 ==0): # condition
        print(i)
        break      # stoping while fisrt found
    i += 1
```

**Exercise 4:** Write a minimal placeholder inside a loop using pass (e.g., iterate 3 times).

```python
i = 0
while i < 3:
    pass
    i += 1
    print(i)
```

# else clause with loops

**Exercise 1:** Use a for-else: loop through 1..5 and print 'no break triggered' in else.

```python
for i in range (1,6):
    print(i)
else:
    print("No break triggered")
```

**Exercise 2:** Search numbers 2..10 for a multiple of 11; if none found, print 'none' from else.

```python
for i in range(2,11):
    if i % 11 == 0:
        print(i)
        break
    else:
        print("none")
```
